```python
from sklearn.model_selection import train_test_split
```

```python
ddri.rental_count.unique()
```

```python
ddri.columns
columns = ['hour', 'cluster', 
           'mapped_dong_code', 'weekday', 'month', 'holiday', 'temperature', 'humidity', 
           'precipitation', 'wind_speed']
```

```python
ddri_morning = ddri[ddri.station_group == '아침 도착 업무 집중형']
ddri_morning[columns]
```

```python
feature = ddri_morning[columns]
target = ddri_morning.rental_count
```

```python
train_data, test_data, train_target, test_target = train_test_split(
    feature, target, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(train_data, train_target)
```

```python
rf.score(train_data, train_target)
```

```python
rf.score(test_data, test_target)
```

```python
from lightgbm import LGBMRegressor
```

```python
lgb = LGBMRegressor(max_depth=-1)
lgb.fit(train_data, train_target)
print(lgb.score(train_data, train_target), lgb.score(test_data, test_target))
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor()
xgb.fit(train_data, train_target)
print(xgb.score(train_data, train_target), xgb.score(test_data, test_target))
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
from sklearn.linear_model import LinearRegression
```

```python
estimators = [
    ('rf', RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1)),
    ('lgb', LGBMRegressor(n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42)),
    ('xgb', XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42))
]

# 2차 모델
final_estimator = LinearRegression()

stack_model = StackingRegressor(
    estimators=estimators,
    final_estimator=final_estimator,
    cv=5,
    n_jobs=-1
)

stack_model.fit(train_data, train_target)

print(stack_model.score(train_data, train_target), stack_model.score(test_data, test_target))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

test_pred = stack_model.predict(test_data)

rmse = np.sqrt(mean_squared_error(test_target, test_pred))
mae = mean_absolute_error(test_target, test_pred)
r2 = r2_score(test_target, test_pred)

print('rmse:', rmse, '| mae:', mae, '| r2:', r2)
```
