```python
import pandas as pd

# file name 합치기
from glob import glob
```

```python
ddri_2023 = glob('../3조 공유폴더/2023 강남구 따릉이 이용정보/*.csv')
ddri_2023
```

```python
# 파일 이름 정렬하기
ddri_2023 = sorted(ddri_2023)
ddri_2023
```

```python
for file_name in ddri_2023:
    try:
        tmp = pd.read_csv(file_name, encoding="cp949")
        print("OK:", file_name)
    except Exception as e:
        print("에러:", file_name)
```

```python
tmp_raw = []

for file_name in ddri_2023:
    try:
        tmp = pd.read_csv(file_name, index_col= False, encoding="cp949")
    except:
        try:
            tmp = pd.read_csv(file_name, index_col= False, encoding="utf-8")
        except:
            tmp = pd.read_csv(file_name, index_col= False, encoding="utf-8-sig")

    tmp_raw.append(tmp)

df1 = pd.concat(tmp_raw, ignore_index=True)
df1
```
