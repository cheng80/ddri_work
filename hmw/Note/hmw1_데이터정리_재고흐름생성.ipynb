## 3. 생성 결과 확인

```python
flow = pd.read_csv(DATA_DIR / "station_hour_bike_flow_2023_2025.csv", encoding="utf-8-sig", parse_dates=["time"])
flow.head()
```

```python
flow.groupby("year").agg(
    rows=("station_id", "size"),
    stations=("station_id", "nunique"),
    rental_sum=("rental_count", "sum"),
    return_sum=("return_count", "sum")
)
```
