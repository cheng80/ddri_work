```python
from pathlib import Path

import branca.colormap as cm
import folium
import pandas as pd
import numpy as np
from folium.plugins import HeatMap
from IPython.display import display
```

```python
PROJECT_ROOT = Path.cwd().parents[1]
FEATURE_PATH = PROJECT_ROOT / "hmw" / "Data" / "2023_아침" / "2023_morning.csv"
STATION_PATH = PROJECT_ROOT / "works" / "01_clustering" / "06_data" / "ddri_common_station_master.csv"
OUTPUT_HTML_PATH = PROJECT_ROOT / "hmw" / "Data" / "ddri_morning_map.html"


def build_folium_map() -> None:

    feature_df = pd.read_csv(FEATURE_PATH)
    station_df = pd.read_csv(STATION_PATH).rename(
        columns={"대여소번호": "station_id", "대여소명": "station_name", "위도": "lat", "경도": "lon"}
    )

    plot_df = feature_df.merge(
        station_df[["station_id", "station_name", "lat", "lon"]],
        on="station_id",
        how="left"
    )

    plot_df = plot_df.dropna(subset=["lat", "lon", "count"]).copy()

    center_lat = plot_df["lat"].mean()
    center_lon = plot_df["lon"].mean()

    fmap = folium.Map(
        location=[center_lat, center_lon],
        zoom_start=13,
        tiles="CartoDB positron"
    )

    station_layer = folium.FeatureGroup(name="대여소", show=True)
    heat_layer = folium.FeatureGroup(name="HeatMap", show=True)

    # 컬러맵 생성
    min_val = plot_df["count"].min()
    max_val = plot_df["count"].max()

    colormap = cm.LinearColormap(
        colors=["#2c7bb6", "#abd9e9", "#ffffbf", "#fdae61", "#d7191c"],
        vmin=min_val,
        vmax=max_val
    )

    colormap.caption = "반납횟수"

    # 대여소 점 표시
    for _, row in plot_df.iterrows():

        radius = max(((row["count"])**0.5) * 0.2, 3)

        folium.CircleMarker(
            location=[row["lat"], row["lon"]],
            radius=radius,
            color="white",
            weight=0.8,
            fill=True,
            fill_color=colormap(row["count"]),
            fill_opacity=0.85,
            popup=folium.Popup(
                f"""
                <b>대여소명</b>: {row['station_name']}<br>
                <b>반납횟수</b>: {row['count']:.2f}
                """,
                max_width=260
            ),
            tooltip=f"{row['station_name']} / 반납횟수 {row['count']:.1f}"
        ).add_to(station_layer)

    # HeatMap
    heat_data = plot_df[["lat", "lon", "count"]].values.tolist()

    HeatMap(
        heat_data,
        radius=20,
        blur=16,
        min_opacity=0.3,
        max_zoom=16
    ).add_to(heat_layer)

    station_layer.add_to(fmap)
    heat_layer.add_to(fmap)

    colormap.add_to(fmap)

    # 범례
    legend_html = """
    <div style="
        position: fixed;
        bottom: 30px;
        left: 30px;
        width: 230px;
        z-index: 9999;
        background: white;
        border: 2px solid #cbd5e1;
        border-radius: 8px;
        padding: 12px;
        font-size: 13px;
    ">
        <b>범례</b><br>
        색상 : 반납횟수<br>
        파랑 → 적음<br>
        빨강 → 많음<br>
        점 크기 = 반납횟수
    </div>
    """

    fmap.get_root().html.add_child(folium.Element(legend_html))

    # 제목
    title_html = """
    <div style="
        position: fixed;
        top: 15px;
        left: 50%;
        transform: translateX(-50%);
        z-index: 9999;
        background: white;
        padding: 10px 18px;
        border-radius: 10px;
        border: 1px solid #cbd5e1;
        font-size: 18px;
        font-weight: 700;
    ">
        2023년 아침 강남구 따릉이 반납 지도
    </div>
    """

    fmap.get_root().html.add_child(folium.Element(title_html))

    folium.LayerControl(collapsed=False).add_to(fmap)

    OUTPUT_HTML_PATH.parent.mkdir(parents=True, exist_ok=True)
    fmap.save(OUTPUT_HTML_PATH)

    display(fmap)


if __name__ == "__main__":
    build_folium_map()
```
